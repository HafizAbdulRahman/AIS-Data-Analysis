```python
# Create a geometry column with Point geometries using latitude and longitude
data['geometry'] = data.apply(lambda row: Point(row['Longitude'], row['Latitude']), axis=1)

# Convert the DataFrame to a GeoDataFrame
gdf = gpd.GeoDataFrame(data, geometry='geometry')

# Set a coordinate reference system (CRS), typically WGS84 (EPSG:4326) for latitude and longitude data
gdf.set_crs(epsg=4326, inplace=True)
```

```python
sample_data = data.sample(n=2000, random_state=42)
```

```python
print(sample_data.shape)
```

```python
import folium


latitude, longitude = sample_data['Latitude'].mean(), sample_data['Longitude'].mean()
m = folium.Map(location=[latitude, longitude], zoom_start=6)


for _, row in sample_data.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=row.get("Name", "No Name"),  
        icon=folium.Icon(color="blue", icon="info-sign")
    ).add_to(m)


m
```

```python
import folium
from folium.plugins import FastMarkerCluster
latitude, longitude = sample_data['Latitude'].mean(), sample_data['Longitude'].mean()
m = folium.Map(location=[latitude, longitude], zoom_start=6)
locations = sample_data[['Latitude', 'Longitude']].values.tolist()
FastMarkerCluster(locations).add_to(m)
m
```
